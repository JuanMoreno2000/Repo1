### (The not so recognized effect of luck on success)

A success rate of 35% is quite bad. It's much worse than flipping a coin for every decision.

Let's take $N$ guys/girls this bad (let's say, in investing), and see how far they make it "in life".

```python
from random import random
import matplotlib.pyplot as plt

N = 1000
success_rate = .35
population = [i for i in range(N)]
```

```python
coo = [(i/N**.5, i%N**.5) for i in range(N)]
dead = []
horizon = range(5)

for year in horizon:
    plt.figure()
    for guy in population:
        if guy not in dead:
            if random() >= success_rate:
                dead.append(guy)
                plt.plot(coo[guy][0], coo[guy][1], 'o', color='grey')
            else:
                plt.plot(coo[guy][0], coo[guy][1], 'o',  color='green')
    plt.title("YEAR "+str(year))
    plt.show()
    
print("We have "+str(N-len(dead))+" people in our 'Walking Dead' set")
print("(i.e., 'unreliable' actors that happened to survive)")
```

## The other ~5 guys make it to Forbes

**The morale of the story is**: luck just helped a few bad-performing people survive in spite of their poor performance/skills. They "should" be broke by now, but they are still walking around. We'll see them giving "expert advice" in magazines (since people ignore the effect of luck and explain success based on skill or other factors).
